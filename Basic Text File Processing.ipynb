```python
#Reading the File
file_path = 'customer_feedback.txt'

#Counting Feedback Entries
keywords = ["good", "bad", "excellent", "poor"]
keyword_counts = {keyword: 0 for keyword in keywords}  # Initialize counters for each keyword

try:
    # Open the file and read all feedback entries
    with open(file_path, 'r', encoding='utf-8') as file:
        feedback_entries = file.readlines()

    total_feedback_entries = len(feedback_entries)  # Total number of feedback entries
    print(f"Total feedback entries: {total_feedback_entries}")

    #Generating a Summary Report
    with open('summary_report.txt', 'w', encoding='utf-8') as report_file:
        # Write header to the report
        report_file.write("--- Summary Report ---\n")
        report_file.write(f"Total feedback entries: {total_feedback_entries}\n")
        report_file.write("Keyword occurrences:\n")

        #Extracting Keywords
        for i, feedback in enumerate(feedback_entries, start=1):
            feedback_cleaned = feedback.strip().lower()  # Clean and lowercase the feedback
            words = feedback_cleaned.split()  # Split the feedback into words

            found_keywords = []  # Track keywords found in this entry

            #Counting Keyword Occurrences
            for word in words:
                if word in keyword_counts:
                    keyword_counts[word] += 1  # Increment the keyword count
                    if word not in found_keywords:
                        found_keywords.append(word)

            #display feedback and matched keywords
            print(f"\nFeedback {i}: {feedback.strip()}")
            if found_keywords:
                print(f"Matched Keywords: {', '.join(found_keywords)}")
            else:
                print("No keywords found.")

        # Write the keyword counts to the report file
        for keyword, count in keyword_counts.items():
            report_file.write(f"{keyword.capitalize()}: {count} times\n")

    #Saving the Report
    print("\nSummary report saved to 'summary_report.txt'.")

except FileNotFoundError:
    print("Error: The file was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
